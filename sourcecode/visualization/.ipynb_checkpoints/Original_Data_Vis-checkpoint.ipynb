### Visualize outliers

### Identify the outliers with negative times

```python
negative_times_filter = (blue_waters_original_df.POSIX_F_FASTEST_RANK_TIME < 0) | (
            blue_waters_original_df.POSIX_F_SLOWEST_RANK_TIME < 0) | (blue_waters_original_df.POSIX_F_META_TIME < 0) | (
                                    blue_waters_original_df.POSIX_TOTAL_TIME < 0) | (
                                    blue_waters_original_df.bandwidth < 0)

blue_waters_posix_negative_outliers_df = blue_waters_original_df.loc[negative_times_filter == True]
blue_waters_posix_negative_outliers_df
```

```python
blue_waters_iqr_filtering_df = blue_waters_original_df.loc[negative_times_filter == False]
```

### Apply IQR to the rest (same as performed on the dataset used for models)

```python
bandwidth_q1 = blue_waters_iqr_filtering_df.bandwidth.quantile(0.25)
bandwidth_q3 = blue_waters_iqr_filtering_df.bandwidth.quantile(0.75)
bandwidth_iqr = bandwidth_q3 - bandwidth_q1
bandwidth_iqr
```

```python
bandwidth_q1
```

```python
bandwidth_q3
```

```python
iqr_filter = (blue_waters_iqr_filtering_df.bandwidth >= bandwidth_q1 - 1.5 * bandwidth_iqr) & (
            blue_waters_iqr_filtering_df.bandwidth <= bandwidth_q3 + 1.5 * bandwidth_iqr)

bandwidth_outliers = blue_waters_iqr_filtering_df.loc[iqr_filter == False]
bandwidth_outliers
```

```python
blue_waters_no_outliers_df = blue_waters_iqr_filtering_df.loc[iqr_filter == True]
blue_waters_no_outliers_df["bandwidth"].describe()
```
